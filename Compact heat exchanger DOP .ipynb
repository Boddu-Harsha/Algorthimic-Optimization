**Mechanical Design**

**Minimum Fin thickness(t) = (P/nf.σf).(ph/ph-dh)**

P = relevant stream design pressure, Pa
=1.5 times design pressures

nf = number of fins per metre, 1/m

ph = pitch of holes in corrugation if fin is perforated, m

dhole = diameter of holes if fin is perforated

σf = allowable tensile stress for fin material at design temperature, Pa

**Side bar and header minimum thickness**

tb=b*((1.25*Pm)/σb)

σb = allowable tensile stress for bar material at design temperature
=275 × 10^6 Pa

P = the relevant stream design pressure (Pa)
=9.1 × 10^5 Pa

**b**b = gap between sheets (m)
=0.00635m

```python
b_gas=0.00635    #m
σb = 275000000   #Pa
pm=float(1.5*pin_air) 

A=float((1.25*pm)/σb)
tb=float((b_gas*1000)*(math.sqrt(A)))

#Headers minimum thickness#
R=1.145                     #m
S=275000000                 #Pa
E=0.85
P_AIR=float(1.5*pin_air)
P_GAS=float(1.5*pin_gas)

t_air=float((P_AIR*R*1000)/((2*S*E)-(0.2*P_AIR))) #mm
t_gas=float((P_GAS*R*1000)/((2*S*E)-(0.2*P_GAS))) #mm
print(tb,t_air,t_gas)
```
